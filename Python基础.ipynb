**基本数据类型**

```python
#数值型
ival=1561 
fval=3.14
fval2=6.78e-5
cval=3+2j
#可以使用整数的除法得到浮点数
fval3=3/2 #此处与c/c++不同
```

```python
print(ival,fval,fval2,cval,fval3)
```

```python
#逻辑型
flag1=True
flag2=False
#在python中逻辑型可以看作是整数型的子类，二者兼容
print(flag1+1) #此处与java不同
```

```python
#字符串，'',"",''''''
a = 'one way of writing a string'
b = "another way"
c = """
This is a longer string that
spans multiple lines
"""
print(a,b)
print(c,c.count('\n'),"个换行符")
```

```python
#转义字符
str1="12\n13"
str2="12\\n13"
str3=r'this\has\no\special\characters'
print(str,str2,sep="\n")
print(str3)
```

```python
#格式化
#%形式
print('the number is %-10.2f'%(3.14))
#format形式（索引，键，序号）
print("the first number is {0:f},the second number is {1:d}".format(3.14,125))
print("the first number is {first:5.2f},the second number is {second:3d}".format(first=3.14,second=125))
num=[3.14,125]
print("the first number is {0[0]:<5.2f},the second number is {0[1]:^5d}".format(num))
#f'string形式
print(f"the first number is {3.14:5.2f},the second number is {125:5d}")
```

```python
#类型转换str int float bool
print(type(str(123)),int("001",2),int('123'))
'''
以下会报错
print(int("3.14))
'''
#进制转换
print(f"二进制：{bin(24)},八进制：{oct(24)},十六进制：{hex(24)}")
```

**运算符与表达式**

**算数运算符**

```python
num1=20
num2=30
print(num1+num2,num1-num2,num1*num2,num1/num2,sep=" ")
#整除
print(num2//num1)
#模运算
print(num1%8)
#幂运算，右结合
print(2**2**3)
```

**逻辑运算符**

```python
#not or and
'''
数值型的逻辑运算 
and：有0为0，否则为最后一个非0数；
or ：全0为0，否则为第一个非0数
'''
print(1 and 2 and 3)
print(0 or 1 or 2)
```

**关系运算符**

```python
#python的语法糖
print(5>4>3>2) #相当于5>4 and 4>3 and 3>2 此处与Java不同
```

**赋值运算符**

```python
#多值情况
m,n,q=2,3,4
print(m,n,q)
m,n,q=[5,6,7]
print(m,n,q)
#交换两个值
a=3
b=4
a,b=b,a
print(a,b)
'''
以下错误
a=(b=c=2)
'''
```

**其他**

```python
#成员运算符 用于序列（字符串，列表，集合，元组，字典）
print(1 in [1,2,3])
#身份运算符
num3=2
num4=num3
print(num3 is num4)
#id()
print(id(num3))
#type()
print(type(num3))
#isinstance()
print(isinstance(num3,int))
```
