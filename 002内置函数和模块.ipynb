**字符串处理函数**

```python
#不可变数据类型的操作都是非原地操作
s="This is a great day"
#大小写转换
print(s.lower(),s.upper(),s.title())
#查找
print(s.find('i'),s.index('i'))
#替换
print(s.replace('i','a',1))
#统计
print(s.count('a'))
print(len(s))
#合并与拆分
arr=s.split(" ")
print(arr)
print(",".join(arr))
```

**random模块**


```python
import random as ran
```

```python
print(f"random()0-1之间的随机浮点数{ran.random()},randint()范围内的随机整数{ran.randint(1,11)},randrange()生成序列中的随机整数{ran.randrange(0,10,2)}")
```

```python
ran.seed(10)#设置随机种子便于结果复现
seq=[1,3,5,7,9]
#shuffle()将一个序列打乱
ran.shuffle(seq)
print(seq)
```

```python
#生成指定范围指定个数的随机整数
#1.可重复
arr1=[ran.randint(0,10) for i in range(10)]#注意range函数是[start,end)
#2.不可重复
arr2=ran.sample(range(10),10)#从seq中选取n个不重复的数
s=set()
while len(s)!=10:
    s.add(ran.randint(0,10)) #注意randint函数是[start,end]
print(arr1)
print(arr2)
print(s)
```

**time模块**

```python
import time
from datetime import datetime
```

```python
t=time.time()#t是一个时间戳
print(t)
tmtup=time.localtime()#tmtup时间元组
print(tmtup)
print(time.localtime(t))#时间戳---->时间元组
print(time.mktime(tmtup))#时间元组--->时间戳
```

```python
#格式化输出
print(time.strftime("%Y-%m-%d %H:%M:%S",tmtup))#将一个时间元组，格式化输出
print(time.strptime("2025-12-11","%Y-%m-%d"))#将一个日期字符串，转化为时间元组
```

```python
from datetime import datetime
```

```python
dt=datetime.today()#datatime对象
print(datetime.strftime(dt,"%Y-%m-%d %H:%M:%S"))#使用datetime类的方法
print(dt.strftime("%Y-%m-%d %H:%M:%S"))#使用对象的方法
print(f"年：{dt.year},月：{dt.month},日：{dt.day},时：{dt.hour},分：{dt.minute}，秒：{dt.second}")#对象的其他方法
```
