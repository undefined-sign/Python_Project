```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
```

**处理缺失数据(na not available)**

```python
'''
在许多数据分析工作中，缺失数据是经常发生的。
pandas的目标之一就是尽量轻松地处理缺失数据。
例如，pandas对象的所有描述性统计默认都不包括缺失数据

'''
'''
缺失数据在pandas中呈现的方式有些不完美，但对于大多数用户可以保证功能正常。
对于数值数据，pandas使用浮点值NaN（Not a Number）表示缺失数据。
我们称其为哨兵值，可以方便的检测出来
'''
string_date=pd.Series(['aardvark', 'artichoke', np.nan, 'avocado'])
print(string_date)
print(string_date.isnull())
#Python内置的None值在对象数组中也可以作为NA
string_date[0]=None
print(string_date.isnull())
```

**过滤缺失数据**

```python
#使用dropna，对于一个Series，dropna返回一个仅含非空数据和索引值的Series
from numpy import nan as NA
data=pd.Series([1, NA, 3.5, NA, 7])
print(data.dropna())
#使用布尔索引
print(data[data.notnull()])
```

```python
#而对于DataFrame对象，事情就有点复杂了。你可能希望丢弃全NA或含有NA的行或列。
#dropna默认丢弃任何含有缺失值的行
data=pd.DataFrame([[1., 6.5, 3.], [1., NA, NA],
                   [NA, NA, NA], [NA, 6.5, 3.]])
data[4]=NA
print(f'原dataframe对象：\n{data}')
print('-----------------------')
print(data.dropna())
print('----------传入how=all将只丢弃全为NA的行-------------')
print(data.dropna(how='all'))
print('----------传入axis=1将只丢弃对应的列-------------')
print(data.dropna(axis=1,how='all'))
```

```python
#另一个滤除DataFrame行的问题涉及时间序列数据。假设你只想留下一部分观测数据，可以用thresh参数实现此目的
df=pd.DataFrame(np.random.randn(7,3))
df.iloc[:4,1]=NA
df.iloc[:2,2]=NA
print(f'原dataframe对象：\n{df}')
print('----------使用dropna函数-------------')
print(df.dropna())
print('----------用thresh参数-------------')
print(df.dropna(thresh=2))#当某一行有两个NA时过滤掉该行
```

**填充缺失数据**

```python
'''
你可能不想滤除缺失数据（有可能会丢弃跟它有关的其他数据），而是希望通过其他方式填补那些“空洞”。
对于大多数情况而言，fillna方法是最主要的函数。
通过一个常数调用fillna就会将缺失值替换为那个常数值
'''
print(df)
```

```python
print('''--------fillna() Must specify a fill 'value' or 'method'---------''')
print(df.fillna(0))
print('---------通过一个字典调用fillna，就可以实现对不同的列填充不同的值------------')
print(df.fillna({1:0.5,2:0}))
```

```python
print('---------fillna默认会返回新的对象,原来的df对象不变-----------')
print(df)
print('------------使用inplace参数可以实现原地修改-----------------')
df_=df.fillna(0,inplace=True)
print(df)
```

```python
df=pd.DataFrame(np.random.randn(6,3))
df.iloc[2:,1]=NA
df.iloc[4:,2]=NA
print(df)
```

```python
print('------method属性用于指定填充方向---------')
print(df.fillna(method='ffill'))#等价于df.ffill()
print('------limit参数限制连续填充的数量--------')
print(df.fillna(method='ffill',limit=2))
```

```python
print('------使用特征值进行填充--------')
data = pd.Series([1., NA, 3.5, NA, 7])
print(f'原始数据:\n{data}')
print(data.fillna(data.mean()))
```

## 数据转换

**移除重复数据**

```python
data = pd.DataFrame({'k1': ['one', 'two'] * 3 + ['two'],
                     'k2': [1, 1, 2, 3, 3, 4, 4]})
print(data)
```

```python
print('----DataFrame的duplicated方法返回一个布尔型Series，表示各行是否是重复行（前面出现过的行)----')
print(data.duplicated())
print('------------dorp_duplicates,去除重复的行--------------')
print(data.drop_duplicates())
'''
这两个方法默认会判断全部列，你也可以指定部分列进行重复项判断。假设我们还有一列值，且只希望根据k1列过滤重复项
'''
print('---------dorp_duplicates[`k1`]按照k1列进行去重-----------')
print(data.drop_duplicates(['k1']))
print('---------keep=last用于保留最后一个重复项-----------')
print(data.drop_duplicates(['k1'],keep='last'))
```
