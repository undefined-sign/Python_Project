```python
'''杨辉三角'''
list1=[[0 for j in range(i+1)] for i in range(10)]#初始化二维列表
for i in range(10):
    for j in range(i+1):
        if j==0 or i==j:
            list1[i][j]=1
        else:
            list1[i][j]=list1[i-1][j]+list1[i-1][j-1]

for i in range(10):
    for j in range(i+1):
        print(list1[i][j],end=" ")
    print() 
```

```python
'''斐波那契'''
list2=[0 for i in range(10)]
list2[0]=1
list2[1]=1
for i in range(2,10):
    list2[i]=list2[i-1]+list2[i-2]
print(f"斐波那契:{list2}")
```

```python
'''进制转换'''
num=eval(input("input a number:"))
b=''
while num!=0:
    s=num%2
    num=num//2
    b=str(s)+b

print(b)
```

```python
'''素数判断'''
n=0
sum=0
for i in range(2,100):
    for j in range(2,i):
        if i%j==0:
            break
    if j>=i-1:        #此处注意为大于等于号
        n=n+1
        sum=sum+i
        if n%5==0:
            print(i)
        else:
            print(i,end="\t")
print(sum)
```

```python
'''闰年判断'''
year=eval(input("input a year"))
if (year%4==0 and year%100!=0)or year%400==0:
    print(f"{year}是闰年")
else:
    print(f"{year}不是闰年")
```

```python
'''九九乘法表'''
for i in range(1,10):
    for j in range(1,i+1):
        print(f"{j}*{i}={i*j}",end="\n" if i==j else "\t")
```

```python
'''一百个铜钱买了一百只鸡，其中公鸡一只5钱、母鸡一只3钱，小鸡一钱3只，问一百只鸡中公鸡、母鸡、小鸡各多少）'''
for i in range(21):
    for j in range(34):
        for k in range(100):
            if i+j+k==100 and (i*5+j*3+k/3)==100:
                print(f"公鸡{i}只,母鸡{j}只,小鸡{k}只")
```

```python
'''
字典的一个典型应用
统计字符数目
'''
str='''Weathered stone paths meander through 
the quiet garden where sunlight filters softly 
between ancient trees and low, fragrant shrubs.
A small fountain murmurs nearby, scattering silver 
ripples across a shallow basin while birds flit among branches, 
their calls brief and bright. Old wooden benches invite visitors to pause, 
breathe deeply, and watch clouds drift lazily above distant rooftops. 
Time seems to slow here, offering a gentle retreat from bustling streets and hurried schedules. 
People arrive expecting little and often leave with renewed calm, quiet smiles, 
and a renewed sense of careful, grateful attention to simple pleasures and peaceful evening reflections.'''

dt=dict()

for s in str:
    dt[s]=dt.get(s,0)+1 
n=0
for k,v in dt.items():
    n=n+1
    if k=='\n':
        print(f'\\n--{v}',end="\t" if n%8!=0 else "\n")
    else:
        print(f'{k}--{v}',end="\t" if n%8!=0 else "\n")
```
